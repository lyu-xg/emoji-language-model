```python
from magpie import MagpieModel
import pandas as pd
import numpy as np
df = pd.read_csv('emoji_hack_word_embed.csv')
data_folder = 'data/'
```

```python
df.drop('embedding', inplace=True, axis=1)
```

```python
magpie = MagpieModel()
```

```python
import os


def format_convert(df=df, data_folder=data_folder):
    os.system('rm -rf {}; mkdir {}'.format(data_folder, data_folder))
    for i,(_,x,y) in enumerate(df.values):
        open('{}{}.lab'.format(data_folder,i),'w').write(y.strip().replace('_','\n'))
        open("{}{}.txt".format(data_folder,i),'w').write(' '.join(eval(x.strip())))
    print('csv convert to werid magpie format... done.')
    
```

```python
format_convert()
```

```python
magpie.init_word_vectors(data_folder, vec_dim=100)
```

```python
def get_all_labels(df=df):
    return list(set(each_y for _,x,y in df.values for each_y in y.strip().split('_')))
```

```python
magpie.train(data_folder, get_all_labels(), test_ratio=0.2, nb_epochs=3000)
```

```python
magpie.predict_from_text("rockstar")
```

```python
magpie.save_word2vec_model('save/embeddings')
```
